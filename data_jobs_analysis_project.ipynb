```python
import pandas as pd

data = {
    "Job_Title": [
        "Data Analyst", "Data Scientist", "Data Engineer",
        "Junior Data Analyst", "Business Analyst",
        "Data Analyst", "Data Engineer", "Data Scientist"
    ],
    "Company": [
        "Company A", "Company B", "Company C",
        "Company D", "Company E",
        "Company F", "Company G", "Company H"
    ],
    "Skill_Required": [
        "SQL, Python, Excel",
        "Python, Machine Learning, SQL",
        "Python, SQL, Cloud",
        "Excel, SQL, Power BI",
        "Excel, SQL, Communication",
        "Python, SQL",
        "Cloud, SQL, Python",
        "Machine Learning, Python"
    ],
    "Experience_Level": [
        "Entry", "Senior", "Mid",
        "Entry", "Mid",
        "Entry", "Mid", "Senior"
    ]
}

df = pd.DataFrame(data)
df
```

```python
df["Job_Title"].value_counts()
```

```python
df["Skill_Required"].value_counts()
```

```python
df["Experience_Level"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Job_Title"].value_counts().plot(kind="bar")
plt.title("Demand for Data Roles")
plt.xlabel("Job Role")
plt.ylabel("Count")
plt.show()
```

## Key Insights

- Data Analyst roles appear frequently, showing strong demand for entry-level positions.
- SQL and Python are the most common skills across data roles.
- Entry-level roles still require strong technical skills like SQL, Excel, and Python.
- Data Engineering roles often require cloud and programming skills.
